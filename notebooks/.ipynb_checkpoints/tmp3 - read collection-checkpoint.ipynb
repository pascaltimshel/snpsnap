```python
i = 1
parrent_snp = df_matrix.index[i] # type --> string
match_snps = df_matrix.ix[i] # series
set_idx = df_matrix.columns.values # better than #set_idx = range(1,len(df_matrix.columns)+1)
df_container = pd.DataFrame(set_idx, columns=['set'])
df_container.ix[:,'input_snp'] = parrent_snp
df_container.head()
```
