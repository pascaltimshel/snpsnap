```python
store = pd.HDFStore(file_db, 'r')
df = store.get('dummy')
store.close()
```

```python
rsIDs = df.ix[:,'rsID']
rsIDs.head()
```

```python
rsIDs_valid = {}
rsIDs_chrpos = {}
rsIDs_other = {}

#for rsID in rsIDs[:100]:
for rsID in rsIDs:
    if rsID.startswith("rs"):
        rsIDs_valid[rsID] = 1
    elif ":" in rsID:
        rsIDs_chrpos[rsID] = 1
    else:
        rsIDs_other[rsID] = 1
```
