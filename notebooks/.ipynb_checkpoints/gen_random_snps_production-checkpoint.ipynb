## Read data

```python
#file_db = "/Users/pascaltimshel/snpsnap/data/step3/1KG_snpsnap_production_v1/ld0.5/ld0.5_db.h5"
file_db = "/Users/pascaltimshel/p_snpsnap/data/step3/1KG_snpsnap_production_v1_uncompressed_incl_rsID/ld0.5/ld0.5_db.h5"
store = pd.HDFStore(file_db, 'r')
df = store.select('dummy') # read data frame
store.close()
```

```python
## Generate random SNPs
n_sample = 100
np.random.seed(1)
#type(df.index.values) #==>numpy.ndarray
rand_idx = np.random.choice(df.index.values, size=n_sample, replace=False, p=None) # sample uniformly
df_rand_samp = df.ix[rand_idx] # new data frame

df.ix[[1, 1, 1]].head()
```

```python
### TESTING reindexing - MAY BE DELETED
df_rand_samp.reindex(index=np.random.permutation(df_rand_samp.index)) 
```

## Write to file

```python
### NEW VERSION - writes using .to_csv(). INCLUDES rsID!
filename = "/Users/pascaltimshel/sample_%drandSNPs.list" % n_sample
df_rand_samp.to_csv(filename, sep='\t', index=True, columns=['rsID'], header=False, mode='w')
```
